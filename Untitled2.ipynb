```python
# ===============================================================
# 🚀 Refined SBERT Model for Water Footprint Search (WandB Disabled)
# ===============================================================

# ✅ Mount Drive
from google.colab import drive
drive.mount("/content/drive")

# ✅ Imports
import os
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer, InputExample, losses, util
from torch.utils.data import DataLoader
import numpy as np
import faiss

# ✅ Disable WandB (prevents API key issues)
os.environ["WANDB_DISABLED"] = "true"
os.environ["WANDB_MODE"] = "offline"

# ✅ Load Dataset
data_path = "/content/drive/MyDrive/WaterFootprintData/maharashtra_water_footprint_ai_generated.csv"
df = pd.read_csv(data_path)
print("Dataset Loaded ✅")
print("Rows:", len(df))
print(df.head())

# ✅ Columns for Natural Statements
CROP_COL = "Crop"
DIST_COL = "District"
WF_COL = "Total_Water_Footprint(L_per_kg)"

# ✅ Generate Natural Sentences for Embeddings
sentences = []
for i, row in df.iterrows():
    crop = str(row[CROP_COL])
    district = str(row[DIST_COL])
    water = str(row[WF_COL])

    sentences.append(f"{crop} grown in {district} requires {water} liters of water per kg.")
    sentences.append(f"Water footprint of {crop} in {district} is {water} L/kg.")

print("✅ Generated sentences:", len(sentences))

# ✅ Create Dummy Identity Training Pairs for fine-tuning (Siamese Objective)
train_examples = [InputExample(texts=[s, s]) for s in sentences]

# ✅ DataLoader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)
train_loss = losses.MultipleNegativesRankingLoss(model=None)  # set later

# ✅ Load SBERT Model
MODEL_NAME = "sentence-transformers/all-MiniLM-L6-v2"
model = SentenceTransformer(MODEL_NAME)

train_loss.model = model  # attach model for loss

# ✅ Fine-Tune (2 Epochs)
EPOCHS = 2
MODEL_SAVE_DIR = "/content/drive/MyDrive/Model_training/wf_sbert_refined"

print("\n🚀 Fine-tuning started...")
model.fit(train_objectives=[(train_dataloader, train_loss)],
          epochs=EPOCHS,
          show_progress_bar=True)

# ✅ Save Model
model.save(MODEL_SAVE_DIR)
print(f"✅ Model saved at: {MODEL_SAVE_DIR}")

# ✅ Encode All Sentences → Build FAISS Index for Fast Search
print("✅ Creating Embeddings...")
embeddings = model.encode(sentences, convert_to_numpy=True)

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

# ✅ Save FAISS index & sentences mapping
faiss.write_index(index, f"{MODEL_SAVE_DIR}/faiss_index.bin")
pd.DataFrame({"sentence": sentences}).to_csv(f"{MODEL_SAVE_DIR}/sentences.csv", index=False)

print("✅ FAISS index saved!")

# ===============================================================
# 🔍 Query Function
# ===============================================================
def search_water_footprint(query, top_k=3):
    print(f"\n🔹 Query: {query}\n")

    query_emb = model.encode(query, convert_to_numpy=True)
    D, I = index.search(np.array([query_emb]), top_k)

    for rank, idx in enumerate(I[0]):
        sentence = sentences[idx]
        print(f"Rank {rank+1}: {sentence}")
        print(f"Score: {1-D[0][rank]:.4f}\n")  # Higher score = better match

# ===============================================================
# ✅ Test Query
# ===============================================================
search_water_footprint("What is the water footprint of wheat in Nagpur?")
```

```python
# ================================================================
# ✅ Maharashtra Water Footprint Chatbot (Corrected Column Names)
# ================================================================

# --- Step 1: Mount Google Drive ---
from google.colab import drive
drive.mount("/content/drive")

# --- Step 2: Import dependencies ---
import os
import re
import pandas as pd
import numpy as np
import torch
from sentence_transformers import SentenceTransformer, util
import faiss
import warnings

warnings.filterwarnings("ignore")

# ✅ Setup Path
dataset_path = "/content/drive/MyDrive/WaterFootprintData/Maharashtra_WaterFootprint_AI.csv"

# ✅ Load Dataset
df = pd.read_csv(dataset_path)
df.drop_duplicates(inplace=True)

# ✅ FIX — Correct column names
req_cols = ["District", "Crop", "Total_Water_Footprint(L_per_kg)"]
missing_cols = [c for c in req_cols if c not in df.columns]

if missing_cols:
    raise Exception(f"❌ Missing Required Columns in CSV: {missing_cols}")

df.dropna(subset=req_cols, inplace=True)

print("✅ Dataset Loaded — Rows:", len(df))
print("✅ Columns:", df.columns.tolist())

# ✅ Create sentence field
def make_sentence(row):
    return f"Water footprint of {row['Crop']} in {row['District']} is {row['Total_Water_Footprint(L_per_kg)']} L/kg."

df["sentence"] = df.apply(make_sentence, axis=1)

# ✅ Load Sentence-BERT
model = SentenceTransformer("all-MiniLM-L6-v2")

# ✅ Encode + FAISS Index
corpus_embeddings = model.encode(df["sentence"].tolist(), show_progress_bar=True)
corpus_embeddings = np.array(corpus_embeddings).astype("float32")

index = faiss.IndexFlatL2(corpus_embeddings.shape[1])
index.add(corpus_embeddings)

# ✅ Entity lists
valid_crops = sorted(df["Crop"].unique())
valid_districts = sorted(df["District"].unique())


# ✅ Extract Crop & District
def extract_entities_from_query(query):
    q = query.lower()
    crop, district = None, None

    for c in valid_crops:
        if c.lower() in q:
            crop = c
            break

    for d in valid_districts:
        if d.lower() in q:
            district = d
            break

    return crop, district


# ✅ Main Query Function
def answer_query(query, top_k=5):
    crop, district = extract_entities_from_query(query)
    print(f"🔍 Parsed Entities: Crop={crop}, District={district}")

    query_embed = model.encode([query]).astype("float32")
    scores, results = index.search(query_embed, top_k)

    results = results[0]
    scores = scores[0]

    filtered = []
    for idx, score in zip(results, scores):
        row = df.iloc[idx]
        if crop and row["Crop"] != crop:
            continue
        if district and row["District"] != district:
            continue
        filtered.append((row, score))

    if not filtered:
        filtered = [(df.iloc[idx], scores[i]) for i, idx in enumerate(results)]

    print(f"\n🔹 Query: {query}\n")
    for row, score in filtered:
        print(f"✅ {row['sentence']}  (Score={score:.4f})")


# ✅ TEST
answer_query("What is the water footprint of grapes in Nashik?")
```
