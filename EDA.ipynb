```python
#  Maharashtra Water Footprint — Dataset Overview

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ✅ Step 1: Load dataset
# (Upload the CSV in Colab before running this)
df = pd.read_csv("/content/drive/MyDrive/WaterFootprintData/Maharashtra_WaterFootprint_AI.csv")

# ✅ Step 2: Dataset info
print(" Total Records:", len(df))
print(" Columns:", list(df.columns))
print("\nDataset Sources: ICAR, NABARD, RKVY, Mekonnen & Hoekstra (2020) — *for presentation context*\n")

# ✅ Step 3: Unique counts
print(" Total Crops:", df['Crop'].nunique())
print(" Total Regions/Districts:", df['District'].nunique())
print(" Irrigation Types:", df['Irrigation_Type'].unique())
print(" Categories:", df['Category'].unique())

# ✅ Step 4: Mean & Median of Total Water Footprint
mean_val = df['Total_Water_Footprint(L_per_kg)'].mean()
median_val = df['Total_Water_Footprint(L_per_kg)'].median()

print(f"\n Mean Water Footprint: {mean_val:.2f} L/kg")
print(f" Median Water Footprint: {median_val:.2f} L/kg")

# ✅ Step 5: Visualization — Crop Count
plt.figure(figsize=(10,5))
sns.countplot(y='Crop', data=df, order=df['Crop'].value_counts().index, palette='Blues_r')
plt.title("Crop-wise Record Count")
plt.xlabel("Number of Records")
plt.ylabel("Crop")
plt.tight_layout()
plt.show()

# ✅ Step 6: Visualization — Irrigation Types
plt.figure(figsize=(6,6))
df['Irrigation_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("Blues"))
plt.title("Distribution of Irrigation Types")
plt.ylabel("")
plt.show()

# ✅ Step 7: Visualization — Category Count
plt.figure(figsize=(6,5))
sns.countplot(x='Category', data=df, palette='Blues')
plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ✅ Step 8: Visualization — District Count
plt.figure(figsize=(10,4))
sns.countplot(y='District', data=df, order=df['District'].value_counts().index, palette='Blues_r')
plt.title("District-wise Record Count")
plt.tight_layout()
plt.show()
```
